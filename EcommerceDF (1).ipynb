```python
# Box plots for 'Quantity' and 'UnitPrice'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Quantity'])
plt.title('Box Plot of Quantity')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['UnitPrice'])
plt.title('Box Plot of UnitPrice')
plt.show()
```

```python
# Distribution plots for 'Quantity' and 'UnitPrice'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Quantity'], bins=10, kde=True)
plt.title('Distribution of Quantity')
plt.subplot(1, 2, 2)
sns.histplot(df['UnitPrice'], bins=10, kde=True)
plt.title('Distribution of UnitPrice')
plt.show()
```

```python
# Count plot for 'Country'
plt.figure(figsize=(16, 6))
sns.countplot(x='Country', data=df)
plt.title('Count of Transactions by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```
